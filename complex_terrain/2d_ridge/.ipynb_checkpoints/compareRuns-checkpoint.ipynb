### Comparison of nalu-wind data using the new wall-stress model with the experimental data of Cao & Tamura 2006

```python
%%capture
# Import libraries
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
rcParams.update({'figure.autolayout': True})
```

```python
#Define the 2D ridge
h=0.04;
b=0.1;
N=200;
x=np.linspace(-2.5*h,2.5*h,N)
zw=h*np.cos(np.pi*x/(2*b))**2.
```

```python
# Load experimental data
Um250=np.genfromtxt('exp_data/2D_ridge_U_m2.50.csv',delimiter=',')
Um125=np.genfromtxt('exp_data/2D_ridge_U_m1.25.csv',delimiter=',')
U0=np.genfromtxt('exp_data/2D_ridge_U_m0.00.csv',delimiter=',')
Up125=np.genfromtxt('exp_data/2D_ridge_U_p1.25.csv',delimiter=',')
Up250=np.genfromtxt('exp_data/2D_ridge_U_p2.50.csv',delimiter=',')
Up375=np.genfromtxt('exp_data/2D_ridge_U_p3.75.csv',delimiter=',')
Up500=np.genfromtxt('exp_data/2D_ridge_U_p5.00.csv',delimiter=',')
Up625=np.genfromtxt('exp_data/2D_ridge_U_p6.25.csv',delimiter=',')
Up750=np.genfromtxt('exp_data/2D_ridge_U_p7.50.csv',delimiter=',')

u2m250=np.genfromtxt('exp_data/2D_ridge_u2_m2.50.csv',delimiter=',')
u2m125=np.genfromtxt('exp_data/2D_ridge_u2_m1.25.csv',delimiter=',')
u20=np.genfromtxt('exp_data/2D_ridge_u2_0.csv',delimiter=',')
u2p125=np.genfromtxt('exp_data/2D_ridge_u2_p1.25.csv',delimiter=',')
u2p250=np.genfromtxt('exp_data/2D_ridge_u2_p2.50.csv',delimiter=',')
u2p375=np.genfromtxt('exp_data/2D_ridge_u2_p3.75.csv',delimiter=',')
u2p500=np.genfromtxt('exp_data/2D_ridge_u2_p5.00.csv',delimiter=',')
u2p625=np.genfromtxt('exp_data/2D_ridge_u2_p6.25.csv',delimiter=',')
u2p750=np.genfromtxt('exp_data/2D_ridge_u2_p7.50.csv',delimiter=',')
```

```python
markS=10
Every=1
plt.figure(1,figsize=(10,8))
ax1=plt.subplot(211)
#plot the 2D ridge
ax1.plot(x/h,zw/h,'-k')

#plot experimental data
ax1.plot(Um250[:,0],Um250[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(-2.50*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax1.plot(Um125[:,0],Um125[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(-1.25*np.linspace(1,1,10),np.linspace(0.5,4.1,10),'--k')
ax1.plot(U0[:,0],U0[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(0*np.linspace(1,1,10),np.linspace(1,4.1,10),'--k')
ax1.plot(Up125[:,0],Up125[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(1.25*np.linspace(1,1,10),np.linspace(0.5,4.1,10),'--k')
ax1.plot(Up250[:,0],Up250[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(2.5*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax1.plot(Up375[:,0],Up375[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(3.75*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax1.plot(Up500[:,0],Up500[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(5.0*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax1.plot(Up625[:,0],Up625[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(6.25*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax1.plot(Up750[:,0],Up750[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(7.50*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')

#legend and axis configuration
ax1.set_xlabel(r'$\langle u \rangle/U_\infty + x/h$',fontsize=22)
ax1.set_ylabel(r'$z/h$',fontsize=22)
ax1.set_yticks([0,1,2.0,3,4])
ax1.set_xticks([-2.5,-1.25,0,1.25,2.5,3.75,5,6.25,7.5,8.75])
ax1.tick_params(labelsize=18)
ax1.set_xlim(-2.6,8.8)
ax1.set_ylim(0,4.1)

ax2=plt.subplot(212)
#plot the 2D ridge
ax2.plot(x/h,zw/h,'-k')

#plot experimental data
ax2.plot(u2m250[:,0],u2m250[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(-2.50*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax2.plot(u2m125[:,0],u2m125[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(-1.25*np.linspace(1,1,10),np.linspace(0.5,4.1,10),'--k')
ax2.plot(u20[:,0],u20[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(0*np.linspace(1,1,10),np.linspace(1,4.1,10),'--k')
ax2.plot(u2p125[:,0],u2p125[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(1.25*np.linspace(1,1,10),np.linspace(0.5,4.1,10),'--k')
ax2.plot(u2p250[:,0],u2p250[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(2.5*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax2.plot(u2p375[:,0],u2p375[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(3.75*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax2.plot(u2p500[:,0],u2p500[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(5*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax2.plot(u2p625[:,0],u2p625[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(6.25*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')
ax2.plot(u2p750[:,0],u2p750[:,1],'or',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(7.50*np.linspace(1,1,10),np.linspace(0,4.1,10),'--k')


#legend and axis configuration
ax2.set_xlabel(r'5 $\sigma_u/U_\infty + x/h$',fontsize=22)
ax2.set_ylabel(r'$z/h$',fontsize=22)
ax2.set_yticks([0,1,2.0,3,4])
ax2.set_xticks([-2.5,-1.25,0,1.25,2.5,3.75,5,6.25,7.5,8.75])
ax2.tick_params(labelsize=18)
ax2.set_xlim(-2.6,8.8)
ax2.set_ylim(0,4.1)
```
