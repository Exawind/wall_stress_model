### Comparison of nalu-wind data with Sullivan et al. 2008

Data digitized from Sullivan et al. 2008 paper

```python
%%capture
# Import libraries
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
rcParams.update({'figure.autolayout': True})
```

```python
# Peter Sullivan data
sully_path='sullivan_data/'

class profile_sullivan:
    def __init__(self,Filename):
        DATA=np.genfromtxt(Filename,delimiter=',',skip_header=1)
        self.z=DATA[:,0]
        self.U=DATA[:,1]
        self.V=DATA[:,2]
        self.u2=DATA[:,3]
        self.v2=DATA[:,4]
        self.w2=DATA[:,5]

NoWaves=profile_sullivan(sully_path+'NoWaves.dat')
Bumps=profile_sullivan(sully_path+'Bumps.dat')
WindFollowing=profile_sullivan(sully_path+'WindFollowing.dat')
WindOpposing=profile_sullivan(sully_path+'WindOpposing.dat')

```

```python
Ug=5
Vg=0.
zi=400.
markS=7
Every=7
plt.figure(1,figsize=(10,7))
ax1=plt.subplot(121)
ax1.plot(NoWaves.U,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(Bumps.U,Bumps.z,'-sk',markerfacecolor='none',markersize=markS,markevery=Every,label='Stationary (PS)')
ax1.plot(WindFollowing.U,WindFollowing.z,'-ob',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind following (PS)')
ax1.plot(WindOpposing.U,WindOpposing.z,'-^k',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind opposing (PS)')
ax1.set_ylabel(r'$z/z_i$',fontsize=22)
ax1.set_xlabel(r'$\langle u \rangle/U_g$',fontsize=22)
ax1.set_xticks([0,.5,1.0])
ax1.set_yticks([0.5,1])
ax1.tick_params(labelsize=18)
ax1.set_xlim(0.,1.2)
ax1.set_ylim(0,1)
ax1.legend(fontsize=11,frameon=False)
ax2=plt.subplot(122)
ax2.plot(NoWaves.V,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(Bumps.V,Bumps.z,'-sk',markerfacecolor='none',markersize=markS,markevery=Every,label='Stationary (PS)')
ax2.plot(WindFollowing.V,WindFollowing.z,'-ob',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind following (PS)')
ax2.plot(WindOpposing.V,WindOpposing.z,'-^k',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind opposing (PS)')
ax2.set_ylabel(r'$z/z_i$',fontsize=22)
ax2.set_xlabel(r'$\langle v \rangle/U_g$',fontsize=22)
ax2.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax2.set_yticks([0.5,1])
ax2.tick_params(labelsize=18)
ax2.set_xlim(-0.2,0.25)
ax2.set_ylim(0,1)
```

```python
markS=7
Every=2
plt.figure(2,figsize=(15,5))
ax1=plt.subplot(131)
ax1.semilogy(NoWaves.u2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.semilogy(Bumps.u2,Bumps.z,'-sk',markerfacecolor='none',markersize=markS,markevery=Every,label='Stationary (PS)')
ax1.semilogy(WindFollowing.u2,WindFollowing.z,'-ob',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind following (PS)')
ax1.semilogy(WindOpposing.u2,WindOpposing.z,'-^k',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind opposing (PS)')
ax1.set_ylabel(r'$z/z_i$',fontsize=22)
ax1.set_xlabel(r'$\langle u^2 \rangle$',fontsize=22)
#ax1.set_xticks([0,.5,1.0])
ax1.set_yticks([0.5,1])
ax1.tick_params(labelsize=18)
#ax1.set_xlim(0.,1.2)
ax1.set_ylim(0.005,1.1)
ax1.legend(fontsize=11,frameon=False)
ax2=plt.subplot(132)
ax2.semilogy(NoWaves.v2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.semilogy(Bumps.v2,Bumps.z,'-sk',markerfacecolor='none',markersize=markS,markevery=Every,label='Stationary (PS)')
ax2.semilogy(WindFollowing.v2,WindFollowing.z,'-ob',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind following (PS)')
ax2.semilogy(WindOpposing.v2,WindOpposing.z,'-^k',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind opposing (PS)')
ax2.set_ylabel(r'$z/z_i$',fontsize=22)
ax2.set_xlabel(r'$\langle v^2 \rangle$',fontsize=22)
#ax2.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax2.set_yticks([0.5,1])
ax2.tick_params(labelsize=18)
#ax2.set_xlim(-0.2,0.25)
ax2.set_ylim(0.005,1.1)
ax3=plt.subplot(133)
ax3.semilogy(NoWaves.w2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax3.semilogy(Bumps.w2,Bumps.z,'-sk',markerfacecolor='none',markersize=markS,markevery=Every,label='Stationary (PS)')
ax3.semilogy(WindFollowing.w2,WindFollowing.z,'-ob',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind following (PS)')
ax3.semilogy(WindOpposing.w2,WindOpposing.z,'-^k',markerfacecolor='none',markersize=markS,markevery=Every,label='Wind opposing (PS)')
ax3.set_ylabel(r'$z/z_i$',fontsize=22)
ax3.set_xlabel(r'$\langle w^2 \rangle$',fontsize=22)
#ax3.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax3.set_yticks([0.5,1])
ax3.tick_params(labelsize=18)
#ax2.set_xlim(-0.2,0.25)
ax3.set_ylim(0.005,1.1)
```

## Compare flat-terrain Nalu wind simulations with PP Sullivan data

```python
Ug=5
Vg=0.
zi=400.

nalu_path='nalu_data/'

class profile_nalu:
    def __init__(self,Filename):
        DATA=np.genfromtxt(Filename,delimiter=' ',skip_header=1)
        self.z=DATA[:,0]
        self.U=DATA[:,1]
        self.V=DATA[:,2]
        self.W=DATA[:,3]
        self.u2=DATA[:,4]
        self.uv=DATA[:,5]
        self.uw=DATA[:,6]
        self.v2=DATA[:,7]
        self.vw=DATA[:,8]
        self.w2=DATA[:,9]


NoWaves_nalu=profile_nalu(nalu_path+'profile.dat')

markS=7
Every=7
plt.figure(1,figsize=(10,4))
ax1=plt.subplot(121)
ax1.plot(NoWaves.U,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.plot(NoWaves_nalu.U/Ug,NoWaves_nalu.z/zi,'-b',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (Nalu-wind)')
ax1.set_ylabel(r'$z/z_i$',fontsize=22)
ax1.set_xlabel(r'$\langle u \rangle/U_g$',fontsize=22)
ax1.set_xticks([0,.5,1.0])
ax1.set_yticks([0.5,1])
ax1.tick_params(labelsize=18)
ax1.set_xlim(0.,1.2)
ax1.set_ylim(0,1)
ax1.legend(fontsize=14,frameon=False)
ax2=plt.subplot(122)
ax2.plot(NoWaves.V,NoWaves.z,'-k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.plot(NoWaves_nalu.V/Ug,NoWaves_nalu.z/zi,'-b',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (Nalu-wind)')
ax2.set_ylabel(r'$z/z_i$',fontsize=22)
ax2.set_xlabel(r'$\langle v \rangle/U_g$',fontsize=22)
ax2.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax2.set_yticks([0.5,1])
ax2.tick_params(labelsize=18)
ax2.set_xlim(-0.2,0.25)
ax2.set_ylim(0,1)


```

```python
plt.figure(2,figsize=(15,5))
ax1=plt.subplot(131)
ax1.semilogy(NoWaves.u2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax1.semilogy(NoWaves_nalu.u2/Ug**2.,NoWaves_nalu.z/zi,'-b',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (Nalu-wind)')
ax1.set_ylabel(r'$z/z_i$',fontsize=22)
ax1.set_xlabel(r'$\langle u^2 \rangle/U_g^2$',fontsize=22)
#ax1.set_xticks([0,.5,1.0])
ax1.set_yticks([0.5,1])
ax1.tick_params(labelsize=18)
#ax1.set_xlim(0.,1.2)
ax1.set_ylim(0.005,1.1)
ax1.legend(fontsize=11,frameon=False)
ax2=plt.subplot(132)
ax2.semilogy(NoWaves.v2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.semilogy(NoWaves_nalu.v2/Ug**2.,NoWaves_nalu.z/zi,'-b',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax2.set_ylabel(r'$z/z_i$',fontsize=22)
ax2.set_xlabel(r'$\langle v^2 \rangle/U_g^2$',fontsize=22)
#ax2.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax2.set_yticks([0.5,1])
ax2.tick_params(labelsize=18)
#ax2.set_xlim(-0.2,0.25)
ax2.set_ylim(0.005,1.1)
ax3=plt.subplot(133)
ax3.semilogy(NoWaves.w2,NoWaves.z,'--k',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax3.semilogy(NoWaves_nalu.w2/Ug**2.,NoWaves_nalu.z/zi,'-b',markerfacecolor='none',markersize=markS,markevery=Every,label='Flat (PS)')
ax3.set_ylabel(r'$z/z_i$',fontsize=22)
ax3.set_xlabel(r'$\langle w^2 \rangle/U_g^2$',fontsize=22)
#ax3.set_xticks([-0.2,-0.1,0.,0.1,0.2])
ax3.set_yticks([0.5,1])
ax3.tick_params(labelsize=18)
#ax2.set_xlim(-0.2,0.25)
ax3.set_ylim(0.005,1.1)
```
